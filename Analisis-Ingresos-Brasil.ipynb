2013

```python
#lee el csv de datos de 2013 y saca las primeras filas para una observacion inicial
df_2013 = pd.read_csv("datos/datos-2013.csv", sep=";", encoding="latin-1")
df_2013.head()
```

```python
#numero de filas y columnas de este dataframe
df_2013.shape
```

Para el año 2013 contamos con 16 columnas y 4498 filas

```python
#se genera una funcion para obtener un reporte para cada dataframe por cada año que de la informacion del numero de nulos, del porcentaje de nulos y del tipo de dato (por cada columna)
def info_reporte(dataframe):
    df_report = pd.DataFrame()
    df_report["Numero_nulos"] = dataframe.isnull().sum()
    df_report["Porcentaje_nulos"] = round((dataframe.isnull().sum()/dataframe.shape[0]*100), 2)
    df_report["Tipo_dato"] = dataframe.dtypes
    return df_report
    
```

```python
#llama a la funcion y la aplica al dataframe
info_reporte(df_2013)
```

```python
#comprueba si hay duplicados por todas las columnas
df_2013.duplicated().sum()
```

```python
#comprueba si hay duplicados por la columna DETALHAMENTO
df_2013.duplicated("DETALHAMENTO").sum()
```

Parece que no hay nulos y que tampoco hay duplicados por todas las columnas. En este caso no contamos con una columna que me identifique cada registro, cada ingreso, por lo tanto es complicado observar si hay duplicados por una columna específica. He pensado que quizás la columna de DETALHAMENTO que te da el detalle del ingreso podría ser como el "identificador unico" en esta tabla pero veo que hay un montón de duplicados también por esta columna asique no es posible. Además, observo que el tipo de dato de las columnas 'VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO', 'PERCENTUAL REALIZADO' noi es correcto ya que aparece como object y debería ser float. La columna DATA LANÇAMENTO también tiene el tipo de dato incorrecto. Esta como tipo object y debería estar en tipo fecha.

```python
#se genera una funcion que va a primero reemplazar comas por puntos y luego va a convertir la columna a tipo de dato decimal
def conversion_tipo_datos(columna):
    return columna.str.replace(",", ".").astype(float)
```

```python
#aplica la funcion a las 4 columnas 'VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO','PERCENTUAL REALIZADO' para cambiar a decimal
df_2013[['VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO',
       'PERCENTUAL REALIZADO']] = df_2013[['VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO',
       'PERCENTUAL REALIZADO']].apply(conversion_tipo_datos)
```

```python
#convierte la columna DATA LANÇAMENTO a tipo fecha
df_2013["DATA LANÇAMENTO"] = pd.to_datetime(df_2013["DATA LANÇAMENTO"], format="%d/%m/%Y")
```

```python
#comprueba que el tipo de dato de todas las columnas es ahora el correcto
df_2013.info()
```

No observo ningún patrón en los nulos, es decir, no es que todos correspondan a un mismo nombre de órgano superior o a una misma categoría económica por ejemplo. Al ser una fecha, sería interesante poder comprobar si los nulos es que corresponden a un mes concreto concreto. He pensado en ordenar la columna de fecha pero los nulos pasan a aparecer al final de la columna por lo que esta práctica no tiene sentido. Por otro lado, se puede comprobar si es que las fechas faltantes cumplen algún patrón. Para ello, voy a generar una funcion que en primer lugar saque todas las fechas comprendidas entre la primera fecha (minimo) y la ultima (maximo) disponibles en 2016. En segundo lugar, saca las fechas que verdaderamente tenemos disponibles en 2016 a través de los valores únicos de esta columna. Por último, hace una comparación entre las fechas completas que deberíamos tener en el dataframe y las fechas que realmente tenemos. Se hace mediante una diferencia que lo que devuelve son aquellas fechas de las que carecemos en la columna.

```python
#se genera la funcion
def comprobacion_fechas(dataframe, columna):
    fecha_completa = pd.date_range(start=dataframe[columna].min(), end = dataframe[columna].max())
    fechas_disponibles = dataframe[columna].unique()
    return fecha_completa.difference(fechas_disponibles)
```

```python
#se invoca a la funcion metiendole como argumentos al dataframe correspondiente y a la columna correspondiente
comprobacion_fechas(df_2016, "DATA LANÇAMENTO")
```

Parece que las fechas faltantes son el 2 de enero, el 3 de enero, el 9 de enero, el 17 de enero, el 24 de enero y el 8 de mayo. Muchas de ellas están concentradas en el mes de enero, y es en este punto en el que habría que tratar de identificar el por qué. Podría estar ocurriendo porque en enero pueden estar ocupados en cerrar las cuentas del ejercicio fiscal anterior, preparación de reportes etc. y esto puede influir en que ciertos ingresos de enero estén quedando pendientes de ser registrados más adelante en el año. 

En este caso, los valores nulos son complicados de rellenar, pues es complicado identificar a cual de esas 6 fechas de registro corresponden. Una posible aproximación podría ser a partir de la media o mediana de las columnas numéricas para cada mes. Es decir, si por ejemplo el valor realizado tuviese una mediana diferente entre meses, podríamos fijarnos en el valor que toma esta columna de valor realizado para cada registro concreto donde la fecha de registro es un nulo y observar a que mediana de que mes se parece más, y entonces rellenar por lo menos el mes. 

```python
#calculo de la mediana para cada mes de las 4 columnas numericas disponibles
df_2016.groupby(df_2016["DATA LANÇAMENTO"].dt.month_name())[['VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO',
       'PERCENTUAL REALIZADO']].median()
```

```python
#comprobar si dentro de los nulos la columna valor realizado toma algun valor en algun registro
df_nulos_2016[df_nulos_2016["VALOR REALIZADO"] != 0]
```

Se puede observar como la mediana de valor realizado (ya que el resto son ceros) es relativamente parecida entre los meses por lo que esta imputación para los nulos se haría bastante complicada. Además, para todos los registros donde fecha de registro es nula, valor realizado es 0 por lo que rellenar los nulos es una tarea bastante complicada. Por ello, la decisión final es mantener los nulos como están y trabajar con ellos. El % de nulos es mínimo. Eliminarlos no es una opción ya que contienen información en el resto de columnas que puede ser interesante para el análisis de las otras columnas.
