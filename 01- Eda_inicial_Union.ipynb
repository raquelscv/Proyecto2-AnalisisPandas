### 2013

```python
#lee el csv de datos de 2013 y saca las primeras filas para una observacion inicial
df_2013 = pd.read_csv("datos/datos-2013.csv", sep=";", encoding="latin-1")
df_2013.head()
```

```python
#saca el numero de filas y columnas de este dataframe
df_2013.shape
```

Para el año 2013 contamos con 16 columnas y 4498 filas

```python
#se aplica al dataframe la funcion de info_reporte
se.info_reporte(df_2013)
```

```python
#comprueba si hay duplicados por todas las columnas
df_2013.duplicated().sum()
```

```python
#comprueba si hay duplicados por la columna DETALHAMENTO
df_2013.duplicated("DETALHAMENTO").sum()
```

Parece que no hay nulos y que tampoco hay duplicados por todas las columnas. En este caso no contamos con una columna que me identifique cada registro, cada ingreso, por lo tanto es complicado observar si hay duplicados por una columna específica. He pensado que quizás la columna de DETALHAMENTO que te da el detalle del ingreso podría ser como el "identificador unico" en esta tabla pero veo que hay un montón de duplicados también por esta columna asique no es posible. Además, observo que el tipo de dato de las columnas 'VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO', 'PERCENTUAL REALIZADO' noi es correcto ya que aparece como object y debería ser float. La columna DATA LANÇAMENTO también tiene el tipo de dato incorrecto. Esta como tipo object y debería estar en tipo fecha

```python
#se genera una funcion que va a primero reemplazar comas por puntos y luego va a convertir la columna a tipo de dato decimal
def conversion_tipo_datos(columna):
    return columna.str.replace(",", ".").astype(float)
```

```python
#aplica la funcion conversion_tipo_dato a las 4 columnas 'VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO','PERCENTUAL REALIZADO' para cambiar a decimal
df_2013[['VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO',
       'PERCENTUAL REALIZADO']] = df_2013[['VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO',
       'PERCENTUAL REALIZADO']].apply(conversion_tipo_datos)
```

```python
#convierte la columna DATA LANÇAMENTO a tipo fecha
df_2013["DATA LANÇAMENTO"] = pd.to_datetime(df_2013["DATA LANÇAMENTO"], format="%d/%m/%Y")
```

```python
#comprueba que el tipo de dato de todas las columnas es ahora el correcto
df_2013.info()
```

El análisis exploratorio de las variables numéricas y categóricas una a una lo voy a hacer más detallado cuando ya haya concatenado todos los dataframes individuales, pero para cada uno, voy a sacar los descriptivos de las variables simplemente para observar si hay algo de un año en concreto y una variable concreta que me sorprenda, o que sea atípico y haya que tener en cuenta

```python
#devuelve los descriptivos de las variables numericas
df_2013.describe().T
```

```python
#devuelve los descriptivos de las variables categoricas
df_2013.describe(include="object").T
```

De primeras, hay varias cosas que me sorprenden. En primer lugar, que solo se disponga de una fecha de registro en este año (el 31 de diciembre). En segundo lugar, que haya valores negativos en la columna de valor realizado y en tercer lugar la cantidad de ceros observados en columnas como valor previsto actualizado, valor lanzado o percentual realizado

```python
#devuelve el registro donde el valor realizado toma el valor mínimo
df_2013[df_2013["VALOR REALIZADO"] == -276398836.04]
```

```python
#devuelve los registros donde valor realizado es negativo
df_2013[df_2013["VALOR REALIZADO"] < 0].head()
```

```python
#devuelve los registros donde PERCENTUAL REALIZADO es negativo 
df_2013[df_2013["PERCENTUAL REALIZADO"] <0].head()
```

Parece que hay más de un registro donde el valor realizado es negativo por lo que considero que no ha sido un error en la introducción de los datos, sino que tiene que tener alguna explicación. Estos valores negativos pueden corresponder a ajustes administrativos, devoluciones, compensaciones tributarias o pérdidas financieras en la ejecución de ingresos. Cada caso variará en función de la naturaleza del ingreso. Por ejemplo, podría ocurrir que en el caso de las multas (que por lo que veo hay varias indicadas en el detalle de los ingresos) se hubiese inicialmente registrado un ingreso positivo por una multa emitida, pero que a posteriori la multa se revisara y se anulara por ejemplo, y que ahora el gobierno este registrando un ingreso negativo con el fin de reflejar esa anulación o devolución del dinero que ya fue registrado. Los valores negativos en la columna del percentual tienen sentido en aquellos registros cuyo valor realizado es negativo ya que la columna del percentual es calculada como (valor realizado/ valor previsto actualizado)*100 y si se preveía un ingreso positivo pero finalmente el recaudado ha sido negativo, el percentual es lógico que salga negativo

No observo ningún patrón en los nulos, es decir, no es que todos correspondan a un mismo nombre de órgano superior o a una misma categoría económica por ejemplo. Al ser una fecha, sería interesante poder comprobar si los nulos es que corresponden a un mes concreto concreto. Por otro lado, se puede comprobar si es que las fechas faltantes cumplen algún patrón. Para ello, voy a generar una funcion que en primer lugar saque todas las fechas comprendidas entre la primera fecha (minimo) y la ultima (maximo) disponibles en 2016. En segundo lugar, saca las fechas que verdaderamente tenemos disponibles en 2016 a través de los valores únicos de esta columna. Por último, hace una comparación entre las fechas completas que deberíamos tener en el dataframe y las fechas que realmente tenemos. Se hace mediante una diferencia que lo que devuelve son aquellas fechas de las que carecemos en la columna

```python
#saca los estadisticos de DATA LANÇAMENTO para verificar cual es el minimo y el maximo 
df_2016["DATA LANÇAMENTO"].describe().T
```

```python
#se invoca a la funcion comprobacion_fechas metiendole como argumentos al dataframe correspondiente y a la columna correspondiente
se.comprobacion_fechas(df_2016, "DATA LANÇAMENTO")
```

Parece que las fechas faltantes son el 2 de enero, el 3 de enero, el 9 de enero, el 17 de enero, el 24 de enero y el 8 de mayo. Muchas de ellas están concentradas en el mes de enero, y es en este punto en el que habría que tratar de identificar el por qué. Podría estar ocurriendo porque en enero pueden estar ocupados en cerrar las cuentas del ejercicio fiscal anterior, preparación de reportes etc. y esto puede influir en que ciertos ingresos de enero estén quedando pendientes de ser registrados más adelante en el año. Además, en este caso, el mínimo y el máximo de la fecha de registro es el 1 de enero y el 31 de diciembre por lo que se están verificando la existencia en la columna de todas las fechas de un año. Si por ejemplo el mínimo fuese en marzo o el máximo fuese en noviembre, esta comprobación no tendría demasiado sentido ya que la diferencia de las fechas podría salir vacía si las fechas faltantes se concentrasen por ejemplo en febrero o en diciembre. No las sacaría, pero en este caso se comprueba correctamente

En este caso, los valores nulos son complicados de rellenar, pues es complicado identificar a cual de esas 6 fechas de registro corresponden. Una posible aproximación podría ser a partir de la media o mediana de las columnas numéricas para cada mes. Es decir, si por ejemplo el valor realizado tuviese una mediana diferente entre meses, podríamos fijarnos en el valor que toma esta columna de valor realizado para cada registro concreto donde la fecha de registro es un nulo y observar a que mediana de que mes se parece más, y entonces rellenar por lo menos el mes

```python
#calcula la mediana para cada mes de las 4 columnas numericas disponibles
df_2016.groupby(df_2016["DATA LANÇAMENTO"].dt.month_name())[['VALOR PREVISTO ATUALIZADO', 'VALOR LANÇADO', 'VALOR REALIZADO',
       'PERCENTUAL REALIZADO']].median()
```

```python
#comprueba si dentro de los nulos la columna valor realizado toma algun valor en algun registro
df_nulos_2016[df_nulos_2016["VALOR REALIZADO"] != 0]
```

Se puede observar como la mediana de valor realizado (ya que el resto son ceros) es relativamente parecida entre los meses por lo que esta imputación para los nulos se haría bastante complicada. Además, para todos los registros donde fecha de registro es nula, valor realizado es 0 por lo que rellenar los nulos es una tarea bastante complicada. Por ello, la decisión final es mantener los nulos como están y trabajar con ellos. El % de nulos es mínimo. Eliminarlos no es una opción ya que contienen información en el resto de columnas que puede ser interesante para el análisis de las otras columnas
