```python
#funcion para obtener un reporte que de la informacion del numero de nulos, del porcentaje de nulos y del tipo de dato (por cada columna)
def info_reporte(dataframe):
    df_report = pd.DataFrame()
    df_report["Numero_nulos"] = dataframe.isnull().sum()
    df_report["Porcentaje_nulos"] = round((dataframe.isnull().sum()/dataframe.shape[0]*100), 2)
    df_report["Tipo_dato"] = dataframe.dtypes
    return df_report
```

```python
#filtra por todas las columnas, desde Nome_órgão_superior hasta Detalhamento para que devuelva un ingreso concreto que se ha observado que tiene dato tanto en el valor previsto como en el realizado
df[(df["Nome_órgão_superior"] == "Presidência da república") & (df["Nome_órgão"] == "Fundo de imprensa nacional") & (df["Nome_unidade_gestora"] == "Fundo de imprensa nacional/exec.orc.financ.") & (df["Categoria_econômica"] == "Receitas correntes - intra-orçamentárias") & (df["Origem_receita"] == "Receita industrial") & (df["Espécie_receita"] == "Receita industrial") & (df["Detalhamento"] == "Receita industrial-principal") & (df["Ano_exercício"] == 2016)]
```

Hay 253 registros de este mismo ingreso que lo recauda el mismo órgano. Un registro contiene la estimación (anual, supongo) y una recaudación. El resto de los registros son recaudaciones de ese ingreso concreto en diferentes fechas. Es decir, hay múltiples recaudaciones reales para este ingreso registradas y tan solo hay una estimación que además coincide en la misma fecha que una de las recaudaciones. En este caso, la fecha en la que aparece la previsión es el 21 de enero. Sin embargo, observamos fechas tanto anteriores (como el 15 de enero) como posteriores (13 de julio por ejemplo) de ejecución de este ingreso. Esto creo que puede ocurrir así porque al fin y al cabo la previsión de cada ingreso está hecha anualmente pero es una previsión actualizada, es decir, es una previsión que puede ser modificada o ajustada a lo largo del año por distintos factores (ajuste por las ejecuciones ya registradas en el año, condiciones económicas como nuevas leyes que hagan modificar la previsión). En los registros donde solo disponemos de la previsión, supongo que en este caso la fecha de registro se refiere más bien a la última fecha donde se ha actualizado esa previsión del ingreso. Si en esta misma fecha hay un valor realizado es porque en la misma fecha donde se actualizó por última vez la previsión del ingreso, se ejecutaría ese ingreso (coincidiría la fecha de previsión con una de las fechas de ejecución del ingreso). Por tanto, es coherente y puede pasar que la fecha de previsión sea posterior, o esté entre medias de las fechas de ejecución de ese ingreso

Por lo tanto, para poder analizar las discrepancias entre lo previsto y lo realmente recaudado, la columna de fecha de registro creo que es bastante irrelevante en el análisis. Si disponemos de una previsión que es anual para cada ingreso, habrá que, anualmente, sacar la recaudación de cada ingreso también, para poder comparar la diferencia entre lo que se esperó que se recaudaría y lo que realmente se recaudó. No tiene sentido comparar la previsión anual con una recaudación mensual, por lo tanto, en este caso, el análisis temporal lo veo coherente hacerlo por años, pero no por meses, y es por ello por lo que considero que la fecha de registro es algo irrelevante

```python
#agrupa por las 7 primeras columnas + año y hace la suma del valor previsto y el valor realizado 
df_estimacion_realizado = df.groupby(["Nome_órgão_superior", "Nome_órgão", "Nome_unidade_gestora", "Categoria_econômica", "Origem_receita", "Espécie_receita", "Detalhamento", "Ano_exercício"])[["Valor_realizado", "Valor_previsto_atualizado"]].sum()
df_estimacion_realizado["Percentual_realizado"] = round((df_estimacion_realizado["Valor_realizado"] / df_estimacion_realizado["Valor_previsto_atualizado"])*100,2).replace(np.inf, "Previsión de 0").replace(-np.inf, "Previsión de 0")
df_estimacion_realizado = df_estimacion_realizado.reset_index()
df_estimacion_realizado
```

De esta forma, para 2016 por ejemplo, lo que estoy logrando es tener para cada ingreso recaudado por cada órgano (que es para lo que tenemos la información anual de la previsión), la previsión anual y la recaudación anual (la suma de todos los meses y días del año). En este punto disponemos de algunos ingresos para los que únicamente dispongo del valor de la previsión. En este caso, será porque se preveía recaudar x cantidad, y finalmente se recaudó cero. También, dispongo de algunos ingresos para los que ocurre el caso contrario, y es que se preveía recaudar 0 pero finalmente se recaudó x cantidad. Por último, dispongo de los que tienen valor distinto de cero tanto en la previsión como en la recaudación real final. En los 3 casos es interesante observar esas discrepancias entre lo que se esperaba recaudar y lo realmente recaudado
