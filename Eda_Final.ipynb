Esto creo que puede ocurrir así porque al fin y al cabo la previsión de cada ingreso está acumulada pero es una previsión actualizada, es decir, es una previsión que puede ser modificada o ajustada a lo largo del año por distintos factores (ajuste por las ejecuciones ya registradas en el año, condiciones económicas como nuevas leyes que hagan modificar la previsión). En los registros donde solo disponemos de la previsión, supongo que en este caso la fecha de registro se refiere más bien a la última fecha donde se ha actualizado esa previsión del ingreso. Si en esta misma fecha hay un valor realizado es porque en la misma fecha donde se actualizó por última vez la previsión del ingreso, se ejecutaría ese ingreso (coincidiría la fecha de previsión con una de las fechas de ejecución del ingreso). Por tanto, es coherente y puede pasar que la fecha de previsión sea posterior, o esté entre medias de las fechas de ejecución de ese ingreso

Por lo tanto, para poder analizar las discrepancias entre lo previsto y lo realmente recaudado, la columna de fecha de registro creo que es bastante irrelevante en el análisis. Si disponemos de una previsión que es el acumulado del año para cada ingreso, habrá que, anualmente, sacar la recaudación de cada ingreso también para poder comparar la diferencia entre lo que se esperó que se recaudaría y lo que realmente se recaudó. No tiene sentido comparar la previsión anual con una recaudación mensual, por lo tanto, en este caso, el análisis temporal lo veo coherente hacerlo por años, pero no por meses, y es por ello por lo que considero que la fecha de registro es algo irrelevante. La fecha de registro si puede ser útil para hacer un análisis temporal (por meses) de los ingresos recaudados, pero de las discrepancias entre lo previsto y recaudado no

```python
#agrupa por las 7 primeras columnas + año y hace la suma del valor previsto y el valor realizado y el promedio del percentual
df_estimacion_realizado = df.groupby(["Nome_órgão_superior", "Nome_órgão", "Nome_unidade_gestora", "Categoria_econômica", "Origem_receita", "Espécie_receita", "Detalhamento", "Ano_exercício"]).agg({"Valor_realizado": "sum", "Valor_previsto_atualizado": "sum"})
df_estimacion_realizado = df_estimacion_realizado.reset_index()
df_estimacion_realizado
```

De esta forma, lo que estoy logrando es tener para cada ingreso recaudado por cada órgano, la previsión anual y la recaudación anual (la suma de todos los meses y días del año). En este punto disponemos de algunos ingresos para los que únicamente dispongo del valor de la previsión. En este caso, será porque se preveía recaudar x cantidad, y finalmente se recaudó cero. También, dispongo de algunos ingresos para los que ocurre el caso contrario, y es que se preveía recaudar 0 pero finalmente se recaudó x cantidad. Por último, dispongo de los que tienen valor distinto de cero tanto en la previsión como en la recaudación real final. En los 3 casos es interesante observar esas discrepancias entre lo que se esperaba recaudar y lo realmente recaudado
